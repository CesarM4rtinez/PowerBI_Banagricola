```python
import pandas as pd
import os
import pathlib
from urllib.parse import urlparse
import traceback
from typing import Optional
```

```python
url = "https://ssf.gob.sv/descargas/balances/xls/er_b_0825.xlsx"
header = 5

df = pd.read_excel(url, header=header) # type: ignore
df
```

```python
columna_excluida = "Concepto"

# Seleccionar columnas objetivo
columnas_objetivo = [col for col in df.columns if col != columna_excluida]

# Reemplazar nulos con 0 temporalmente para evaluar condición
df_temp = df[columnas_objetivo].fillna(0)

# Crear condición: todas las columnas objetivo tienen valor 0 (originalmente nulo o cero)
condicion = (df_temp == 0).all(axis=1)

# Filtrar el DataFrame eliminando esas filas
df_filtrado = df[~condicion]

df_filtrado
```

```python
index = [0,1,12,13,32,33,41,43]

limpiar_df = df.drop(index=index)
limpiar_df
```

```python
df_limpio = df_filtrado
df_limpio.to_excel("Dist/estado_resultados_ssf.xlsx", index=False)
```
